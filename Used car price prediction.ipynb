# Replacing 'Null' values by 0

```python
data['Mileage']=data['Mileage'].str.replace('null','0')
data['Power']=data['Power'].str.replace('null','0')
data['Engine']=data['Engine'].str.replace('null','0')
```

# Converting objects into float/int

```python
data['Mileage']=data['Mileage'].astype(float)
data['Engine']=data['Engine'].astype(float)
data['Power']=data['Power'].astype(float)
```

```python
data.dtypes
```

# Convertin 0 to Null to Missing values

```python
data.loc[data.Mileage==0,'Mileage']=np.NaN
data.loc[data.Engine==0,'Engine']=np.NaN
data.loc[data.Power==0,'Power']=np.NaN
```

```python
data.isna().sum()
```

#Filling the missing values

```python
lst=['Mileage','Power','Engine']
for i in lst:
 data[i]=data[i].fillna(data[i].mean())
data['Seats']=data['Seats'].fillna(data['Seats'].mode()[0])
```
